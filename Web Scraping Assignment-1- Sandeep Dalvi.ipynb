# Question 2

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Make a GET request to the URL
url = "https://www.imdb.com/chart/top/"
response = requests.get(url)
```

```python
# Create a BeautifulSoup object from the HTML content
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# extract the data we need
data = []
for movie in soup.find_all("td", class_="titleColumn"):
    name = movie.a.text
    year = movie.span.text[1:5]
    rating = movie.next_sibling.next_sibling.strong.text
    data.append((name, year, rating))
```

```python
# create a data frame from the extracted data
df = pd.DataFrame(data, columns=["Name", "Year", "IMDB Rating"])
```

```python
# display the data frame
print(df.head(50))
```

# Question 4

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Request the website and retrieve the content
url = "https://presidentofindia.nic.in/former-presidents.htm"
res = requests.get(url)
```

```python
# Parse the content using BeautifulSoup
soup = BeautifulSoup(res.text, "html.parser")
```

```python
# Find the list containing the data
list = soup.find("ul", class_="listing cf")
```

```python
data = []
for president in list.find_all("li"):
    name = president.find("h3").text
    term = president.find("span", class_="terms").text
    term = term.replace("Term of Office:", "").strip()
    data.append((name, term))
```

```python
# Create a Pandas dataframe from the data
df = pd.DataFrame(data, columns=["Name", "Term of Office"])
```

```python
# Display the dataframe
print(df.head())
```

# Question 5

a)	Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Send a GET request to the website
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
response = requests.get(url)
```

```python
# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Extract the top 10 ODI teams table
teams_table = soup.find("table", {"class": "table"})
teams_rows = teams_table.find_all("tr")[1:]
```

```python
# Create a list to store the data for each team
teams = []
for row in teams_rows:
    cells = row.find_all("td")
    team = {
        "rank": int(cells[0].text.strip()),
        "team": cells[1].text.strip(),
        "matches": int(cells[2].text.strip().replace(',', '')),
        "points": int(cells[3].text.strip().replace(',', '')),
        "rating": int(cells[4].text.strip().replace(',', ''))
    }
    teams.append(team) 
```

```python
# Convert the list of teams to a Pandas dataframe
teams_df = pd.DataFrame(teams)
```

```python
# Print the dataframes
print("Top 10 ODI Teams:")
print(teams_df)
```
